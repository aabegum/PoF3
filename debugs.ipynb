```python
import pandas as pd
import os

print("\n==============================")
print("CHECKING PoF3 INPUT FILE COLUMNS")
print("==============================\n")

paths = {
    "FAULT DATA": "data/inputs/fault_merged_data.xlsx",
    "HEALTHY DATA": "data/inputs/health_merged_data.xlsx",
    "FAULT EVENTS (after step 1)": "data/intermediate/fault_events_clean.csv",
    "EQUIPMENT MASTER (after step 1)": "data/intermediate/equipment_master.csv",
    "SURVIVAL BASE (after step 1)": "data/intermediate/survival_base.csv",
}

for name, path in paths.items():
    print(f"\n--- {name} ---")
    if not os.path.exists(path):
        print(f"[MISSING] File not found: {path}")
        continue
    try:
        if path.endswith(".xlsx"):
            df = pd.read_excel(path)
        else:
            df = pd.read_csv(path, encoding="utf-8-sig")
    except Exception as e:
        print(f"[ERROR] Could not load: {e}")
        continue

    print(f"Rows: {len(df):,}")
    print(f"Columns ({len(df.columns)}):")
    print(df.columns.tolist())
```
